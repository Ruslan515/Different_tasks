### Избавимся от категориальных признаков

```python
cols_for_encoding = [
    "from",
    "where",
    "time",
    "aircraft_type",
    "service_type"
]

tg_enc_train = TargetEncoder(
    cols=cols_for_encoding,
    smoothing=2,
    min_samples_leaf=4,
)
```

```python
tg_enc_train.fit(X_train, y_train)
X_train_enc = tg_enc_train.transform(X_train)
X_train_enc.head()
```

```python
tg_enc_val =  TargetEncoder(
    cols=cols_for_encoding,
    smoothing=False,    
)
tg_enc_val.fit(X_train, y_train)
X_val_enc = tg_enc_val.transform(X_val)
X_val_enc.head()
```

```python
df_test.head(3)
```

```python
tg_enc_test = tg_enc_train = TargetEncoder(
    cols=cols_for_encoding,
    smoothing=False,    
)
tg_enc_test.fit(X_train, y_train)
X_test_enc = tg_enc_test.transform(df_test)
X_test_enc.head()
```

### используем catboost

```python
def fit_catboost(
    X_train_boost, X_val_boost, 
    y_train, y_test, 
    catboost_params = {}, 
    verbose = 100
):
    learn_pool = Pool(
        X_train_boost, 
        y_train, 
    )
    test_pool = Pool(
        X_val_boost,
        y_test, 
    )
    catboost_default_params = {
        'iterations': 5000,
        'learning_rate': 0.00515,
        'eval_metric': 'RMSE',  # выбрал данную метрику для оценки модели
        # 'task_type': 'GPU',
        'use_best_model': True,
        "early_stopping_rounds":2000        
    }
    catboost_default_params.update(catboost_params)
    
    model = CatBoostRegressor(**catboost_default_params)
    model.fit(learn_pool, eval_set=test_pool, verbose=verbose)
    return model

cat_boost_model = fit_catboost(X_train_enc, X_val_enc, y_train, y_val)
```

```python
y_pred = cat_boost_model.predict(X_test_enc)
y_pred = list(map(round, y_pred))
```

### Посмотрим какие фичи наиболее важные для нашей модели с помощью SHAPE

```python
feature_names = X_train_enc.columns
```

```python
# feature importance for catboost classifier
fi = {n:v for n,v in zip(feature_names,cat_boost_model.feature_importances_.T)}
sorted_fi = sorted(fi.items(), key=operator.itemgetter(1))

plt.rcdefaults()
fig, ax = plt.subplots()

ax.barh(range(len(sorted_fi)), [v[1] for v in sorted_fi], align='center')
ax.set_yticks(range(len(sorted_fi)))
ax.set_yticklabels([v[0] for v in sorted_fi])
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_title('CatBoost feature importance')
```

CatBoost уделяет внимание в первую очередь service_type, where. Данный факт частично расходится с моей логикой - при покупке доп услуг я бы уделил внимание "типу услуг", но куда я лечу для меня не является причиной покупки. Здесь вторым по важности выступает(лично для меня) время в полете.

```python
explainer = shap.TreeExplainer(cat_boost_model)
shap_values = explainer.shap_values(X_train_enc)
shap.summary_plot(shap_values, X_train_enc, feature_names)
```
