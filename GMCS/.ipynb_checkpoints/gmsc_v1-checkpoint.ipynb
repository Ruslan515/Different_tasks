```python
def to_day_week(df):    
    df["day_week"] = df["date"].dt.day_of_week
    df.drop("date", inplace=True, axis=1)
    return df
df_train, df_test = to_day_week(df_train), to_day_week(df_test)
df_train.head()
```

```python
# создадим новый признак - отношение кол-а продоваемых мест к общему кол-у мест
def new_f(df):
    df["percent_services"] = df["num_available"] / df["seats"]
    return df
new_f(df_train).tail()
```

```python
df_train["percent_services"], df_test["percent_services"] = list(
    map(
        lambda x: x["num_available"] / x["seats"], 
        [df_train, df_test]
    )
)
df_train.head()
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    df_train.drop("actual_demand", axis=1),
    df_train["actual_demand"],
    random_state=42,
    test_size=0.2    
)
X_train.head(3)
```

```python
cols_for_encoding = [
    "from",
    "where",
    "time",
    "aircraft_type",
    "service_type"
]

tg_enc_train = TargetEncoder(
    cols=cols_for_encoding,
    smoothing=2,
    min_samples_leaf=4,
)
```

```python
tg_enc_train.fit(X_train, y_train)
X_train_enc = tg_enc_train.transform(X_train)
X_train_enc.head()
```

```python
tg_enc_val =  TargetEncoder(
    cols=cols_for_encoding,
    smoothing=False,    
)
tg_enc_val.fit(X_train, y_train)
X_val_enc = tg_enc_val.transform(X_val)
X_val_enc.head()
```

```python
df_test.head(3)
```

```python
tg_enc_test = tg_enc_train = TargetEncoder(
    cols=cols_for_encoding,
    smoothing=False,    
)
tg_enc_test.fit(X_train, y_train)
X_test_enc = tg_enc_test.transform(df_test)
X_test_enc.head()
```

### используем catboost

```python
def fit_catboost(
    X_train_boost, X_val_boost, 
    y_train, y_test, 
    catboost_params = {}, 
    verbose = 100
):
    learn_pool = Pool(
        X_train_boost, 
        y_train, 
    )
    test_pool = Pool(
        X_val_boost,
        y_test, 
    )
    catboost_default_params = {
        'iterations': 5000,
        'learning_rate': 0.00515,
        'eval_metric': 'RMSE',
        'task_type': 'GPU',
        'use_best_model': True,
        "early_stopping_rounds":2000        
    }
    catboost_default_params.update(catboost_params)
    
    model = CatBoostRegressor(**catboost_default_params)
    model.fit(learn_pool, eval_set=test_pool, verbose=verbose)
    return model

cat_boost_model = fit_catboost(X_train_enc, X_val_enc, y_train, y_val)
```

```python
y_pred = cat_boost_model.predict(X_test_enc)
y_pred = list(map(round, y_pred))
```
