### Найти виджет, который дольше всего загружается на странице pdp и home 

```python
# пошел не очень оптимальным путем выделяя данные в 2 разных фрейма
mask_pdp = (df["page_current"] == "pdp") 
mask_home = (df["page_current"] == "home")
```

```python
df_pdp = df[mask_pdp]#.sort_values("seconds", ascending=False)
df_pdp.head()
```

```python
df_home = df[mask_home]#.sort_values("seconds", ascending=False)
df_home.head()
```

```python
df_e = df_pdp.merge(df_home, left_on="action_widget", right_on= "action_widget")
df_e.head()
```

```python
# наш виджет
df_e.sort_values(["seconds_x", "seconds_y"], ascending=[False, False]).loc[0, "action_widget"]
```
